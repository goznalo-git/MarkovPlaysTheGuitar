# GP5 Data inspection

```python
import guitarpro as gp
```

```python
#song = gp.parse("Data/Mushi-shi-theme.gp5", encoding='UTF-8')
song = gp.parse("Data/Ginpatsu_no_shounen_Killua.gp5", encoding='UTF-8')
#song = gp.parse("Data/PokemonCenter.gp5", encoding='UTF-8')
```

```python
key = None
frac = None

for track in song.tracks:
    
    print(track.name)
    
    for measure in track.measures:
        
        if measure.keySignature.name != key:
            key = measure.keySignature.name
            print('Measure key:', key)
        
        if str(measure.timeSignature.numerator) + "/" + str(measure.timeSignature.denominator.value) != frac:
            frac = str(measure.timeSignature.numerator) + "/" + str(measure.timeSignature.denominator.value)
            print('Measure time signature:', frac)
        
        for voice in measure.voices[:-1]: # there is an extra, empty voice to remove at the end
            
            for beat in voice.beats:
                #print(beat.__dict__.keys())
                #print(beat.duration)
                #print(len(beat.notes))
                #print(beat.effect)
                
                if len(beat.notes) == 0:
                    print("Silence", "--- Duration:", beat.duration.value)
                                
                for note in beat.notes:
                    
                    print("Fret:", note.value, "--- String:", note.string, "--- Duration:", beat.duration.value)
                    #print(note.effect)
                    #print(note)
                print("... BEAT ...")
                
            print("\n** VOICE ***")
                
        print("=== MEASURE ===\n")     
```

Note: be very careful when placing things to be updated inside loops over these things (for instance, counters). Because inside a measure there are voices, which don't appear to add much, but can mess up things.
